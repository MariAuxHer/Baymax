```python
import pandas as pd
import numpy as np
```

```python
'''
Prints out code for separating every column of a dataframe
into individual variables for easy sorting
'''
def DataFrameColumnCodeGenerator(df, comment_top = True, lines_btwn = 0):
    
    '''
    Returns the number of digits in an integer
    '''
    def numDigits(n):
        count=0
        while(n>0):
            count=count+1
            n=n//10
        return count
    
    # grabs a list of column names from the dataframe
    cols_names = df.columns.tolist()
    
    # finds the longest string in the column names
    max_len = -1
    for i in col_names:
        if len(i) > max_len:
            max_len = len(i)

    count = 0
    max_digits = -1
    
    # loops through all column names, printing out the code and comments for each variable
    for col_name in col_names:
        
        # creates lowercase variable names with underscores instead of spaces
        var_name = col_name.replace(" ", "_").lower()


        # places comments above column names
        if comment_top == True:
            print("#", col_name)

        # places comments to the side of column names with even spacing
        else:
            
            # spacing string is created so that all comments are displayed evenly
            spacing = ""
            
            if max_digits == -1:
                max_digits = numDigits(len(col_names))

            current_digits = numDigits(count)
            if count == 0:
                current_digits = 1
            
            # spacing addition for iloc digits
            spacing_add = max_digits - current_digits
            
            total_spacing = (max_len - len(col_name)) + spacing_add

            # creates spacing string for printing
            for j in range(0, total_spacing):
                spacing += " "

            # prints variable and comment (e.g. "col_1  = data.iloc[:, 0 ]      #  Col 1")
            print(var_name, " = data.iloc[:,", count, "]", spacing, " # ", col_name, end='')
            print()
            
            for i in range(0, lines_btwn):
                print()

        count += 1
```

```python
data = pd.read_csv("./NPPES_Data_Dissemination_082123_082723_Weekly/npidata_pfile_20230821-20230827.csv", encoding = "ISO-8859-1", engine='python')
```

```python
DataFrameColumnCodeGenerator(data, comment_top = False, lines_btwn = 0)
```
