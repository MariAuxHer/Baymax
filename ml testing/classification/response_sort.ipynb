```python
import numpy as np
```

```python
def create_individual_files(input_file, output_dir):
    # Open the original text file for reading
    with open(input_file, 'r') as file:
        # Read lines from the file
        lines = file.readlines()

        # Create individual text files for each line
        for i, line in enumerate(lines):
            # Define the name for the individual file
            output_file = f'%s/{i + 1}.txt' % output_dir

            # Write the line to the individual file
            with open(output_file, 'w') as individual_file:
                individual_file.write(line.strip())

            print(f"Created {output_file}")
```

```python
# optometry data conversion

input_file = 'C:/Users/jhuff19/Desktop/Baymax/classification/classification_data/optometry_train.txt'
output_dir = 'C:/Users/jhuff19/Desktop/Baymax/classification/classification_data/train/optometry'

# Call the function to create individual files
create_individual_files(input_file, output_dir)

input_file = 'C:/Users/jhuff19/Desktop/Baymax/classification/classification_data/optometry_test.txt'
output_dir = 'C:/Users/jhuff19/Desktop/Baymax/classification/classification_data/test/optometry'

# Call the function to create individual files
create_individual_files(input_file, output_dir)
```
